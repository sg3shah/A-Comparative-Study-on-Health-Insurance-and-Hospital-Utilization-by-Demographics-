```python
import pandas as pd
import MySQLdb
from sqlalchemy import create_engine
```

```python
# DB Connection
myvars = {}
with open("../risaac-mysql-password") as myfile:
    for line in myfile:
        name, var = line.partition(":")[::2]
        myvars[name.strip()] = var.strip()
```

```python
myvars['DB databasename'] = 'I501saptpurkFall22grp_06_db'
```

```python
conn = MySQLdb.connect(host="localhost", user=myvars['DB username'], passwd=myvars['DB password'], db=myvars['DB databasename'])
cursor = conn.cursor()
```

```python
df_p_demo = pd.read_sql('select * from P_DEMO_preprocess', conn)
df_p_huq = pd.read_sql('select * from P_HUQ_preprocess', conn)
df_p_hiq = pd.read_sql('select * from P_HIQ_preprocess', conn)
```

```python
df_p_huq = df_p_huq.drop("huq_key", axis = 1)
```

```python
df_p_hiq = df_p_hiq.drop("hiq_key", axis = 1)
```

```python
# shape before merge
df_p_demo.shape
```

```python
# merge demographics with hospital utilization dataset
df_p_demo_huq = df_p_demo.merge(df_p_huq, on = "SEQN", how = "inner")
```

```python
# merge data with hospital insurance dataset
df_p_demo_huq_hiq = df_p_demo_huq.merge(df_p_hiq, on = "SEQN", how = "inner")
```
